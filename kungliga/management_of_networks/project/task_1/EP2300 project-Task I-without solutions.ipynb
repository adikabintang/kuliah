# Overview 

Telecom and Internet services are vital to our daily life. Such services involve large and complex software systems that increasingly run on general-purpose paltforms and operating systems. For the proper function of those serives, a service provider performs real-time service assurance which specifies the level of service which customer will receive. In this project we will investigate the quality of service for Video-on-Demand (VoD) service based on the SLA which is expressed as minimum number of frames per secods on the customer terminal. The project focuse on a critical part of service assurance, namely, on the capability of a provider to estimate the service quality based on measurements on the provider's infrustructure. The estimation uses machine-learning techniques, specifically regression and classification. 

# Project Objective and Approach

The objective of this project is to investigate SLA conformance of Video-on-Demand (VoD) service using service metric Y on the client side and device statistic X on the service side. Figure 1 gives the basic configuration of the system we consider [1]. It consists of a client machine that is connected to a server via a network. The client accesses the VoD service that runs on the server. In this setting, device statistics refer to operating-system metrics on the server side, while service metric Y refer to statistics on the client side. Table 1 describes the metrics X and Y used in the project.

![title](overall_framework.png)

Using machine-learning techniques, the problem is to find a function (i.e., a learning model) $M:  X \rightarrow \hat{Y}$, such that $\hat{Y}$ closely approximates Y for a give X. If Y is a real number, the problem is referred to as a regression problem; if Y is a label, the problem is called a classification problem. This project considers both of these problems. To estimate M, measurement pairs (or observations) of the form (X,Y) are needed. A set of measurement pairs is also called a trace, and we use in this project a trace of 3600 observations, collected once every second from running the system over the course of an hour (periodic trace in[2]).

For this project, we say that the VoD service conforms to the SLA at a particular time, if $Y \geq 18 $ frames/second at that time; otherwise, we say that the VoD service violates the SLA.
    
Figure 1 is a simplification of a practical system. It does not take into account network statistics and client device statistics. Also, the service platform is generally much more complex.


The field ID and description of features which are used in this project are as following:

- "cswch/s": Total number of context switches per second.

- "file-nr": Number of file handles used by the system.

- "runq-sz": Run queue length (number of tasks waiting for run time).

- "plist-sz": Number of tasks in the task list.

- "ldavg-1": System load average for the last minute. 

- "totsck": Total number of sockets used by the system.

- "pgfree/s": Number of pages placed on the free list by the system per second.

- "proc/s": Total number of tasks created per second.

- "all_%%usr": Percentage of CPU utilization that occurred while executing at the user level (application).

- "%%memused": Percentage of used memory.

and the service metric Y is:

- "DispFrames": Video Frame Rate
